```python
# importing libraries
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# reading in data
df = pd.read_csv('ipl.csv')
```

```python
df.head()
```

```python
# checking data's information
df.info()
```

```python
#removing unnecessary columns
df.drop(['mid','venue','batsman','bowler','striker','non-striker'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# we will keep only those teams that play regularly in ipls and drops those who appeared in one or two seasons

df['bat_team'].unique()
```

```python
Regular_Teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals','Mumbai Indians','Kings XI Punjab',
       'Royal Challengers Bangalore', 'Delhi Daredevils','Sunrisers Hyderabad']
```

```python
def team_check(team):
    if team in Regular_Teams:
        return True
    else :
        return False
```

```python
df = df[(df['bat_team'].apply(team_check)) & (df['bowl_team'].apply(team_check)) ]
```

```python
print(df['bat_team'].unique())
print(df['bowl_team'].unique())
```

```python
# Now we'll only keep data after 5 overs as later we need atleast 5 overs data to make predictions

df[df['overs']>=5]
```

```python
# Converting date to datetime

df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```
